# Parabolic Mirror

```python
import numpy as np

from optiland import optic
```

```python
# Initialize an optic system
system = optic.Optic()

# add surfaces to the system
system.add_surface(index=0, thickness=np.inf)  # object at infinity

# conic=-1 --> parabolic surface
system.add_surface(
    index=1,
    thickness=-50,
    radius=-100,
    conic=-1,
    material="mirror",
    is_stop=True,
)
system.add_surface(index=2)

# add aperture
system.set_aperture(aperture_type="EPD", value=42)

# add field
system.set_field_type(field_type="angle")
system.add_field(y=0)

# add wavelength
system.add_wavelength(value=0.587, is_primary=True)
```

```python
system.draw(num_rays=4)
```
