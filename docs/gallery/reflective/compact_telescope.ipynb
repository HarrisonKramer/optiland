# Compact Telescope

```python
import numpy as np

from optiland import optic, physical_apertures
```

```python
# Initialize an optic system
system = optic.Optic()

# define physical aperture of first mirror
aperture1 = physical_apertures.RadialAperture(r_min=1.7, r_max=4.2)

# define physical aperture of third mirror
aperture3 = physical_apertures.RadialAperture(r_min=0.65, r_max=1.7)

# add surfaces to the system
system.add_surface(index=0, thickness=np.inf)
system.add_surface(
    index=1,
    thickness=-5.5119,
    radius=-16.4794,
    material="mirror",
    conic=-0.9774823,
    aperture=aperture1,
)
system.add_surface(
    index=2,
    thickness=5.9056,
    radius=np.inf,
    material="mirror",
    is_stop=True,
)
system.add_surface(
    index=3,
    thickness=-5.9056,
    radius=-5.2556,
    material="mirror",
    conic=-0.4605796,
    aperture=aperture3,
)
system.add_surface(index=4, thickness=9.2807, radius=np.inf, material="mirror")
system.add_surface(index=5)

# add aperture
system.set_aperture(aperture_type="imageFNO", value=5.0)

# add field
system.set_field_type(field_type="angle")
system.add_field(y=0)
system.add_field(y=0.5)

# add wavelength
system.add_wavelength(value=0.587, is_primary=True)
```

```python
system.info()
```

```python
system.draw(num_rays=2)
```
