# Starlight Scope Objective

```python
import numpy as np

from optiland import optic, physical_apertures
```

```python
class StarlightScope(optic.Optic):
    """Starlight Scope Objective

    Milton Laikin, Lens Design, 4th ed., CRC Press, 2007, p. 188
    """

    def __init__(self):
        super().__init__()

        # add physical aperture on surface 3
        aperture3 = physical_apertures.RadialAperture(r_max=np.inf, r_min=0.75)

        self.add_surface(index=0, thickness=np.inf)
        self.add_surface(index=1, thickness=0.275, radius=-7.7618, material="BK7")
        self.add_surface(index=2, thickness=1.3662, radius=-16.8186)
        self.add_surface(
            index=3,
            thickness=-1.3662,
            radius=-5.6768,
            material="mirror",
            aperture=aperture3,
        )
        self.add_surface(index=4, thickness=-0.275, radius=-16.8186, material="BK7")
        self.add_surface(
            index=5,
            thickness=0.275,
            radius=-7.7618,
            material="mirror",
            is_stop=True,
        )
        self.add_surface(index=6, thickness=1.3662, radius=-16.8186)
        self.add_surface(index=7, thickness=0.15, radius=-15.1741, material="SF2")
        self.add_surface(index=8, thickness=0.1846, radius=np.inf, material="N-LAK7")
        self.add_surface(index=9, thickness=0.3873, radius=-2.7866)
        self.add_surface(index=10, thickness=0.2469, radius=1.1320, material="N-SK16")
        self.add_surface(index=11, thickness=0.5, radius=1.4144)
        self.add_surface(index=12)

        self.set_aperture(aperture_type="imageFNO", value=1.57)

        self.set_field_type(field_type="angle")
        self.add_field(y=0)
        self.add_field(y=3.75)
        self.add_field(y=7.5)

        self.add_wavelength(value=0.48613270)
        self.add_wavelength(value=0.58756180, is_primary=True)
        self.add_wavelength(value=0.65627250)

        self.update_paraxial()
```

```python
telescope = StarlightScope()
telescope.draw(num_rays=2)
```
