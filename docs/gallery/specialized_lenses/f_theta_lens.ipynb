# F-theta Scan Lens

```python
import numpy as np

from optiland import optic
```

```python
class FTheta(optic.Optic):
    """F-theta lens

    Milton Laikin, Lens Design, 4th ed., CRC Press, 2007, p. 251
    """

    def __init__(self):
        super().__init__()

        self.add_surface(index=0, thickness=np.inf, radius=np.inf)
        self.add_surface(index=1, thickness=5.0, is_stop=True)
        self.add_surface(index=2, radius=-2.2136, thickness=0.3, material="SF57")
        self.add_surface(index=3, radius=-2.6575, thickness=0.02)
        self.add_surface(index=4, radius=-5.5022, thickness=0.5292, material="SF57")
        self.add_surface(index=5, radius=-3.8129, thickness=4.2927)
        self.add_surface(index=6, radius=7.9951, thickness=0.59, material="SF57")
        self.add_surface(index=7, radius=8.3651, thickness=18.0014)
        self.add_surface(index=8)

        # add aperture
        self.set_aperture(aperture_type="EPD", value=1.0)

        # add field
        self.set_field_type(field_type="angle")
        self.add_field(y=0)
        self.add_field(y=5.0)
        self.add_field(y=10.0)
        self.add_field(y=15.0)

        # add wavelength
        self.add_wavelength(value=0.633, is_primary=True)
```

```python
lens = FTheta()
lens.draw()
```
