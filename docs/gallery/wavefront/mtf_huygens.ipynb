# Huygens MTF

```python
from optiland.mtf import HuygensMTF
from optiland.samples.objectives import CookeTriplet
```

```python
lens = CookeTriplet()
_ = lens.draw()
```

```python
mtf = HuygensMTF(
    lens,
    fields="all",
    wavelength="primary",
    num_rays=128,
    image_size=256,
    max_freq="cutoff"
)
_ = mtf.view()
```
