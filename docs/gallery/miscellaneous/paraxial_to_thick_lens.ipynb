# Paraxial to Thick Lens Conversion

Paraxial surfaces can be automatically converted into a thick lens equivalent.

```python
import optiland.backend as be
from optiland.optic import Optic
from optiland.surfaces import convert_to_thick_lens
```

```python
lens = Optic()

# add surfaces
lens.add_surface(index=0, thickness=be.inf)
lens.add_surface(index=1, surface_type='paraxial', thickness=25, f=100)
lens.add_surface(index=2, surface_type='paraxial', thickness=25, f=-50, is_stop=True)
lens.add_surface(index=3, surface_type='paraxial', thickness=100, f=63.5)
lens.add_surface(index=4)

# add aperture
lens.set_aperture(aperture_type='EPD', value=20)

# add field
lens.set_field_type(field_type='angle')
lens.add_field(y=0)
lens.add_field(y=5)

# add wavelength
lens.add_wavelength(value=0.55, is_primary=True)

_ = lens.draw(num_rays=5)
```

```python
lens_new = convert_to_thick_lens(lens)
_ = lens_new.draw(num_rays=5)
```
