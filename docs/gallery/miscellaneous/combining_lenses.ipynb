# Merging Optics

```python
from optiland import optic
```

```python
class Singlet(optic.Optic):
    def __init__(self):
        super().__init__()

        # add surfaces
        self.add_surface(index=0, thickness=50)
        self.add_surface(
            index=1,
            thickness=7,
            radius=43.7354,
            is_stop=True,
            material="N-SF11",
        )
        self.add_surface(index=2, radius=-46.2795, thickness=50)
        self.add_surface(index=3)

        # add aperture
        self.set_aperture(aperture_type="EPD", value=25)

        # add field
        self.set_field_type(field_type="angle")
        self.add_field(y=0)
        # self.add_field(y=3)

        # add wavelength
        self.add_wavelength(value=0.48613270)
        self.add_wavelength(value=0.58756180, is_primary=True)
        self.add_wavelength(value=0.65627250)
```

```python
lens1 = Singlet()
lens2 = Singlet()
```

```python
lens1.draw(num_rays=5)
```

```python
lens_combined = lens1 + lens2  # lens1 + lens 1 is also possible
lens_combined.draw(num_rays=5)
```
