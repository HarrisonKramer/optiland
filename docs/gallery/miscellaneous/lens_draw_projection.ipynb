# 2D Visualization in Non-default Axes

```python
import optiland.backend as be
from optiland.optic import Optic
```

```python
toroidal_lens = Optic()

# add surfaces
toroidal_lens.add_surface(index=0, radius=be.inf, thickness=be.inf)
toroidal_lens.add_surface(
    index=1,
    thickness=7,
    radius_x=30,
    radius_y=50,
    is_stop=True,
    material="N-BK7",
    surface_type="toroidal",
    conic=0.0,
    toroidal_coeffs_poly_y=[],
)
toroidal_lens.add_surface(index=2, thickness=65)
toroidal_lens.add_surface(index=3)

# add aperture
toroidal_lens.set_aperture(aperture_type="EPD", value=20.0)

# add field
toroidal_lens.set_field_type(field_type="angle")
toroidal_lens.add_field(y=0)

# add wavelength
toroidal_lens.add_wavelength(value=0.587, is_primary=True)
```

```python
toroidal_lens.draw(num_rays=10, projection="YZ")  # default
```

Plotting in the `XZ` plane shows the focusing behavior in the XZ sagittal plane.

```python
toroidal_lens.draw(num_rays=10, projection="XZ")
```
