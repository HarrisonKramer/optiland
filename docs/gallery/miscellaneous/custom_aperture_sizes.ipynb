# Specifying Lens Sizes

The physical aperture of a lens can be specified when adding surfaces by providing the `aperture` argument. This specifies the physical _diameter_ of a lens.

```python
import optiland.backend as be
from optiland.optic import Optic
```

```python
class Eyepiece(Optic):
    """Based on US05587837"""

    def __init__(self):
        super().__init__()

        # add surfaces
        self.add_surface(index=0, radius=be.inf, thickness=be.inf)
        self.add_surface(index=1, radius=be.inf, thickness=63.5, is_stop=True)
        self.add_surface(
            index=2, radius=be.inf, thickness=7.62, material="N-LAK10", aperture=55
        )
        self.add_surface(index=3, radius=-62.7223, thickness=1.39954, aperture=55)
        self.add_surface(
            index=4, radius=112.642, thickness=10.16, material="N-LAK10", aperture=55
        )
        self.add_surface(index=5, radius=-86.0407, thickness=2.23774, aperture=55)
        self.add_surface(
            index=6, radius=-58.7746, thickness=3.429, material="SF11", aperture=55
        )
        self.add_surface(
            index=7, radius=47.6019, thickness=12.7, material="N-LAK9", aperture=55
        )
        self.add_surface(index=8, radius=670.814, thickness=0.5588, aperture=55)
        self.add_surface(
            index=9, radius=44.6933, thickness=12.7, material="N-LAK10", aperture=55
        )
        self.add_surface(index=10, radius=-236.547, thickness=0.5588, aperture=55)
        self.add_surface(index=11, radius=27.1511, thickness=7.62, material="SF11")
        self.add_surface(index=12, radius=21.5676, thickness=12.7139)
        self.add_surface(index=13)

        # add aperture
        self.set_aperture(aperture_type="EPD", value=8.0)

        # add field
        self.set_field_type(field_type="angle")
        self.add_field(y=0)
        self.add_field(y=12)
        self.add_field(y=18)

        # add wavelength
        self.add_wavelength(value=0.48613270)
        self.add_wavelength(value=0.58756180, is_primary=True)
        self.add_wavelength(value=0.65627255)
```

```python
lens = Eyepiece()
lens.draw()
```
