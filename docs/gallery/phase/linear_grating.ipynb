## Linear Grating

This notebook demonstrates how to create a surface with a linear grating phase profile.

```python
from optiland.optic import Optic
import optiland.backend as be
from optiland.phase import LinearGratingPhaseProfile
```

```python
optic = Optic()
optic.add_surface(index=0, radius=be.inf, thickness=be.inf)

phase_profile = LinearGratingPhaseProfile(period=3.0, order=1, angle=be.pi/2)

optic.add_surface(
    index=1,
    phase_profile=phase_profile,
    is_stop=True,
    material="air",
    thickness=100.0,
)
optic.add_surface(index=2)

optic.add_wavelength(0.55)
optic.set_field_type("angle")
optic.add_field(0.0)
optic.set_aperture("EPD", 25.0)

_ = optic.draw(wavelengths=[0.45, 0.55, 0.65])
```
