## Radial Phase

This notebook demonstrates how to create a surface with a radial grating phase profile.

```python
from optiland.optic import Optic
import optiland.backend as be
from optiland.phase import RadialPhaseProfile
```

```python
optic = Optic()
optic.add_surface(index=0, radius=be.inf, thickness=be.inf)  # object surface

phase_profile = RadialPhaseProfile(coefficients=[-1e-2, -3e-5])

optic.add_surface(
    index=1,
    phase_profile=phase_profile,
    is_stop=True,
    material="air",
    thickness=120.0,
)
optic.add_surface(index=2)  # image surface

optic.add_wavelength(0.55)
optic.set_field_type("angle")
optic.add_field(0)
optic.set_aperture("EPD", 50.0)

_ = optic.draw(wavelengths=[0.45, 0.55, 0.65])
```
