# Telephoto

```python
import numpy as np

from optiland import optic
```

```python
class Telephoto(optic.Optic):
    """Kingslake Telephoto Lens, Modified by Hopkins

    Warren Smith, Modern Lens Design, 1992, Page 171
    """

    def __init__(self):
        super().__init__()

        self.add_surface(index=0, radius=np.inf, thickness=np.inf)
        self.add_surface(index=1, radius=24.607, thickness=5.08, material="BK7")
        self.add_surface(index=2, radius=-36.347, thickness=1.6, material="F2")
        self.add_surface(index=3, radius=212.138, thickness=12.3)
        self.add_surface(index=4, radius=np.inf, thickness=21.699, is_stop=True)
        self.add_surface(index=5, radius=-14.123, thickness=1.52, material="BK7")
        self.add_surface(index=6, radius=-38.904, thickness=4.8, material="F2")
        self.add_surface(index=7, radius=-25.814, thickness=37.934)
        self.add_surface(index=8)

        self.set_aperture(aperture_type="imageFNO", value=5.6)

        self.set_field_type(field_type="angle")
        self.add_field(y=0)
        self.add_field(y=2.96)
        self.add_field(y=4.19)

        self.add_wavelength(value=0.4861)
        self.add_wavelength(value=0.5876, is_primary=True)
        self.add_wavelength(value=0.6563)
```

```python
lens = Telephoto()
lens.draw()
```
