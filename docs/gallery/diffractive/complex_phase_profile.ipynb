# Custom Phase Grating

This script demonstrates how to create a custom phase profile by subclassing `BasePhase`. This allows for the implementation of arbitrary phase functions on a surface. It creates a simple system with a surface that has a radially varying phase profile and observes its effect on traced rays.

```python
from optiland import optic, phase
import optiland.backend as be
```

```python
class RadialPhase(phase.BasePhase):
    """
    A custom phase profile that varies radially.
    """
    def __init__(self, coefficient):
        super().__init__()
        self.coefficient = coefficient

    def phase_calc(self, x, y):
        r_sq = x**2 + y**2
        return self.coefficient * r_sq

class CustomPhaseSystem(optic.Optic):
    """
    Defines an optical system with a custom phase surface.
    """
    def __init__(self):
        super().__init__()
        # Surface 0: Object plane
        self.add_surface(index=0, z=0)

        # Surface 1: Custom phase surface
        # Small coefficient for radial phase
        custom_phase = RadialPhase(coefficient=1e-3)
        self.add_surface(
            index=1,
            z=50,
            phase_model=custom_phase,
            is_stop=True
        )

        # Surface 2: Image plane
        self.add_surface(index=2, z=100)

        # System aperture and field
        self.set_aperture(aperture_type="EPD", value=20)
        self.set_field_type(field_type="object_height")
        self.add_field(y=0)

        # Wavelength
        self.add_wavelength(value=0.550, is_primary=True)
```

```python
lens = CustomPhaseSystem()
_ = lens.draw(num_rays=11, figsize=(6, 6))
```
