# Pupil Aberration

```python
from optiland.analysis import PupilAberration
from optiland.samples.objectives import DoubleGauss
```

```python
lens = DoubleGauss()
```

```python
lens.draw()
```

```python
pupil_aberration = PupilAberration(lens, fields=[(0, 1.0)])
pupil_aberration.view()
```
