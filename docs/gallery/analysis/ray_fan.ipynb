# Ray Aberration Fan

```python
from optiland.analysis import RayFan
from optiland.samples.objectives import HeliarLens
```

```python
lens = HeliarLens()
```

```python
lens.draw()
```

```python
ray_fan = RayFan(lens, fields=[(0, 1)])
ray_fan.view()
```

```python
ray_fan = RayFan(lens)  # plots all fields by default
ray_fan.view()
```
